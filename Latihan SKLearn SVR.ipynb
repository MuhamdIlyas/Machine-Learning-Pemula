```python
#memanggil dataset dan import library pandas
import pandas as pd

#membaca dataset dan mengubahnya menjadi dataframe
path = "/content/drive/MyDrive/Dataset/Salary_Data1.csv"
data = pd.read_csv(path)

#melihat dataframe
data.head()
```

```python
#melihat info/missing value dataframe
data.info()
```

```python
#memisahkan atribut dan label
import numpy as np

X = data['YearsExperience']
y = data['Salary']

#mengubah bentuk atribut
X = X[:,np.newaxis]
```

```python
#membuat objek SVR
from sklearn.svm import SVR

#membangun model dengan parameter C, gamma, dan kernel
model = SVR(C=1000, gamma=0.005, kernel='rbf')

#melatih model dengan fungdi fit
model.fit(X, y)
```

```python
#memvisualisasikan model SVR
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, model.predict(X))
```
